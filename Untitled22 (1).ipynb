 the code below transform the xml to csv but in the xml, there are some data I don't want to use for example step count or hearthbeat. I just take the calories and date as columns.

```python
import xml.etree.ElementTree as ET
import pandas as pd
from collections import defaultdict

# XML dosyasını içe aktar
tree = ET.parse("Copy of dışa aktarılan.xml")  # Google Colab'a export.xml dosyasını yükle
root = tree.getroot()

daily_calories = defaultdict(float)

for record in root.findall("Record"):
    if record.attrib.get("type") == "HKQuantityTypeIdentifierActiveEnergyBurned":
        date = record.attrib["startDate"].split(" ")[0]
        value = float(record.attrib["value"])
        daily_calories[date] += value

df = pd.DataFrame(daily_calories.items(), columns=["Date", "CaloriesBurned"])
df.to_csv("calories.csv", index=False)
```

In this code, I read the API csv file and seperate column names with comma then there are unnecessary characters. Also,I put df.head() to be sure the code works.

```python
import pandas as pd
import csv

# 1) Dosya adını kendi yüklediğinle değiştir
FNAME = "d30ddef0267f6741e8b52f9e80f8469b (1).csv"

# 2) read_csv'i Python motoru ve QUOTE_NONE ile çağır
df = pd.read_csv(
    FNAME,
    sep=",",
    engine="python",          # bu parametreye izin veren okuma motoru
    quoting=csv.QUOTE_NONE,   # tırnak içindekileri ayırıcı sanma
    skipinitialspace=True     # virgülden sonra boşlukları kırp
)

# 3) Sütun adlarından kalan " işaretlerini kaldır
df.columns = df.columns.str.replace('"','').str.strip()

# 4) Artık gerçek sütun adların var
print(df.columns.tolist())
# -> ['dt', 'dt_iso', 'timezone', 'city_name', ..., 'wind_speed', ...]

# 5) dt sütunundaki tırnakları kaldır ve sonra integer'a ve datetime'a çevir
df["dt"] = df["dt"].str.strip('"') # Tırnakları kaldır
df["dt"] = df["dt"].astype(int)
df["dt"] = pd.to_datetime(df["dt"], unit="s")

# 6) Eğer sadece tarih kısmı gerekiyorsa
df["Date"] = df["dt"].dt.date

# Kontrol:
df.head()
```

```python
# Sayısal olanları dönüştür (sorunlu değerleri atar)
df["temp"] = pd.to_numeric(df["temp"], errors="coerce")
df["wind_speed"] = pd.to_numeric(df["wind_speed"], errors="coerce")
df["rain_1h"] = pd.to_numeric(df["rain_1h"], errors="coerce")
```

In the below, the code calculate the daily mean for temperature and wind speed. Also, the code calculate the total rain.

```python
daily_summary = df.groupby("Date").agg({
    "temp": "mean",
    "wind_speed": "mean",
    "rain_1h": "sum"
}).reset_index()

daily_summary.columns = ["Date", "AvgTemp", "AvgWindSpeed", "TotalRain"]
daily_summary.to_csv("daily_weather.csv", index=False)
daily_summary.head()
```

the code in below, create the merged.csv for temperature, wind speed and rain.

```python
import pandas as pd

# --- 1) Daily Weather dosyasını temizle ---
with open("daily_weather (3).csv", "r", encoding="utf-8") as f:
    raw_w = f.read().splitlines()
clean_w = [ln.strip('"') for ln in raw_w]
with open("daily_weather_clean.csv", "w", encoding="utf-8") as f:
    f.write("\n".join(clean_w))

# --- 2) Calories dosyasını temizle ---
with open("calories.csv", "r", encoding="utf-8") as f:
    raw_c = f.read().splitlines()
clean_c = [ln.strip('"') for ln in raw_c]
with open("calories_clean.csv", "w", encoding="utf-8") as f:
    f.write("\n".join(clean_c))

# --- 3) Temizlenmiş CSV’leri pandas ile oku ---
weather_df  = pd.read_csv("daily_weather_clean.csv", sep=",")
calories_df = pd.read_csv("calories_clean.csv",       sep=",")

# --- 4) Tarih sütunlarını datetime’a çevir ---
weather_df["Date"]  = pd.to_datetime(weather_df["Date"])
calories_df["Date"] = pd.to_datetime(calories_df["Date"])

# --- 5) Birleştir ---
merged_df = pd.merge(
    calories_df,
    weather_df,
    on="Date",
    how="inner"
)

# --- 6) Sonucu kontrol et ve kaydet ---
print(merged_df.head())
merged_df.to_csv("merged_output.csv", index=False)
```

```python
import pandas as pd

# 1) Son 200 günü seç
cutoff = merged_df['Date'].max() - pd.Timedelta(days=200)
df200  = merged_df[ merged_df['Date'] >= cutoff ].copy()

# 2) Outlier kaldırma fonksiyonu
def drop_outliers(df, col):
    q1  = df[col].quantile(0.25)
    q3  = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    # aralığın içinde kalanları döndür
    return df[ df[col].between(lower, upper) ]

# 3) Önce AvgWindSpeed sonra CaloriesBurned’den outlier’ları temizle
df_clean = drop_outliers(df200, 'AvgWindSpeed')
df_clean = drop_outliers(df_clean,  'CaloriesBurned')

# 4) Sonucu kontrol et
print(f"total days number for 200 days with outliers: {len(df200)}")
print(f"total days number for 200 days without outliers: {len(df_clean)}")



# 6) Temizlenmiş veriyi CSV’e kaydet
df_clean.to_csv("merged_last200_no_outliers.csv", index=False)
```

In this result, we see there is so weak correlation between wind speed and tempreture.

```python
import scipy.stats as stats

# 1) X ve Y serilerini al
x = df_clean["AvgWindSpeed"]
y = df_clean["CaloriesBurned"]

# 2) NaN’leri eleyelim
mask = x.notna() & y.notna()
x_clean = x[mask]
y_clean = y[mask]

# 3) Pearson korelasyon testi
r, p = stats.pearsonr(x_clean, y_clean)

print(f"last 200 days (without outliers) Pearson r: {r:.4f}")
print(f"last 200 days (without outliers) p-value: {p:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import t, pearsonr

# 1) Load your cleaned data
df = pd.read_csv("merged_last200_no_outliers.csv", parse_dates=["Date"])
x = df["AvgWindSpeed"].dropna()
y = df["CaloriesBurned"].dropna()

# 2) Compute Pearson correlation and p-value
r, p_value = pearsonr(x, y)
n = len(x)
dfree = n - 2

# 3) Convert r to observed t-statistic
t_obs = r * np.sqrt(dfree / (1 - r**2))

# 4) Prepare null t-distribution
x_t = np.linspace(t.ppf(0.001, dfree), t.ppf(0.999, dfree), 300)
y_t = t.pdf(x_t, dfree)

# 5) Plot
plt.figure(figsize=(8, 4))
plt.plot(x_t, y_t, label=f"t-distribution (df={dfree})")
plt.fill_between(
    x_t, y_t,
    where=np.abs(x_t) >= abs(t_obs),
    color="gray", alpha=0.5,
    label="p-value region"
)
plt.axvline(t_obs, color="red", linestyle="--", label=f"t_obs = {t_obs:.2f}")
plt.title("Null Distribution of t-Statistic\n(Shaded Area = p-value)")
plt.xlabel("t value")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.show()
```
